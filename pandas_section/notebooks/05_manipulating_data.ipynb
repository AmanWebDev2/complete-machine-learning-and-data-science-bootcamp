```python
import requests
def download_car_sales_missing_data():
    response = requests.get("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/refs/heads/master/data/car-sales-missing-data.csv")
    if response.status_code == 200:
        path = f"../data/car-sales-missing.csv"
        with open(path,'wb') as file:
            file.write(response.content)
    else:
        print("failed to download")
download_car_sales_missing_data()        
```

```python
car_sales_missing_df = pd.read_csv("../data/car-sales-missing.csv")
car_sales_missing_df
```
